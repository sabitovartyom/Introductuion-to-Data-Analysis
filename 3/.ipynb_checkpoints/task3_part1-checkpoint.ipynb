**1.** –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ –æ—Ü–µ–Ω–∫—É $\widehat{\theta}$ –º–µ—Ç–æ–¥–æ–º –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –ø–æ —è–≤–Ω–æ–π —Ñ–æ—Ä–º—É–ª–µ.

**2.** –í—ã–ø–∏—à–∏—Ç–µ —Ñ–æ—Ä–º—É–ª—ã –∏—Ç–µ—Ä–∞—Ü–∏–π –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞ (GD) –∏ —Å—Ç–æ—Ö–∞—Å—Ç–∏—á–µ—Å–∫–æ–≥–æ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞ (SGD) –¥–ª—è –ø–æ–∏—Å–∫–∞ $\widehat{\theta}$. –ù–µ –∑–∞–±—É–¥—å—Ç–µ —É–∫–∞–∑–∞—Ç—å, –∫–∞–∫ –≤ SGD –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –±–∞—Ç—á (–Ω–∞–±–æ—Ä –æ–±—ä–µ–∫—Ç–æ–≤ –Ω–∞ –∫–∞–∂–¥–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏).

**3.** –û–ø–∏—à–∏—Ç–µ, —á—Ç–æ –ø—Ä–æ–∏–∑–æ–π–¥–µ—Ç, –µ—Å–ª–∏ –¥–ª—è –Ω–µ–∫–æ—Ç–æ—Ä–æ–≥–æ $i$ –æ—à–∏–±–∫–∞ $\varepsilon_i$ –±—É–¥–µ—Ç –æ—á–µ–Ω—å –±–æ–ª—å—à–æ–π.


*–ó–∞–º–µ—á–∞–Ω–∏—è.*
1. –≠—Ç–æ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è –∑–∞–¥–∞—á–∞, –Ω–µ –ø–µ—Ä–µ–ø—É—Ç–∞–π—Ç–µ üòÑ
2. –í –æ—Ç–≤–µ—Ç–µ –≤–æ –≤—Å–µ—Ö –ø—É–Ω–∫—Ç–∞—Ö –º–æ–≥—É—Ç –æ—Å—Ç–∞–≤–∞—Ç—å—Å—è —Å—É–º–º—ã $n$ —Å–ª–∞–≥–∞–µ–º—ã—Ö, –Ω–æ –Ω–µ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –æ–ø–µ—Ä–∞—Ü–∏–π —Å –≤–µ–∫—Ç–æ—Ä–∞–º–∏ –∏–ª–∏ –º–∞—Ç—Ä–∏—Ü–∞–º–∏.
3. **–ü—Ä–æ—á–∏—Ç–∞–π—Ç–µ —Ç—Ä–µ–±–æ–≤–∞–Ω–∏—è –∫ —Ñ–æ—Ä–º–∞—Ç—É —Ä–µ—à–µ–Ω–∏—è –≤ –ø—Ä–∞–≤–∏–ª–∞—Ö —Å–¥–∞—á–∏ –∑–∞–¥–∞–Ω–∏—è.**

**–†–µ—à–µ–Ω–∏–µ:**

## 1. –û—Ü–µ–Ω–∫–∞ $\widehat{\theta}$ –º–µ—Ç–æ–¥–æ–º –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ (–ú–ù–ö)

–ú–µ—Ç–æ–¥ –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ —Å—Ç—Ä–µ–º–∏—Ç—Å—è –º–∏–Ω–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å —Å—É–º–º—É –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –æ—à–∏–±–æ–∫:

$$L(\theta) = \sum_{i=1}^{n} (Y_i - \theta x_i)^2$$

–ß—Ç–æ–±—ã –Ω–∞–π—Ç–∏ –º–∏–Ω–∏–º—É–º, –≤–æ–∑—å–º–µ–º –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é –ø–æ $\theta$ –∏ –ø—Ä–∏—Ä–∞–≤–Ω—è–µ–º –µ–µ –∫ –Ω—É–ª—é:

$$\frac{\partial L(\theta)}{\partial \theta} = \sum_{i=1}^{n} 2(Y_i - \theta x_i)(-x_i) = 0$$


$$\sum_{i=1}^{n} (Y_i x_i - \theta x_i^2) = 0$$


$$\sum_{i=1}^{n} Y_i x_i = \theta \sum_{i=1}^{n} x_i^2$$

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –æ—Ü–µ–Ω–∫–∞ $\widehat{\theta}$ –º–µ—Ç–æ–¥–æ–º –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤:

$$\widehat{\theta} = \frac{\sum_{i=1}^{n} Y_i x_i}{\sum_{i=1}^{n} x_i^2}$$

## 2. –§–æ—Ä–º—É–ª—ã –∏—Ç–µ—Ä–∞—Ü–∏–π –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞ (GD) –∏ —Å—Ç–æ—Ö–∞—Å—Ç–∏—á–µ—Å–∫–æ–≥–æ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞ (SGD)

### –ì—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π —Å–ø—É—Å–∫ (GD):

GD –∏—Å–ø–æ–ª—å–∑—É–µ—Ç –≤–µ—Å—å –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –≥—Ä–∞–¥–∏–µ–Ω—Ç–∞ –Ω–∞ –∫–∞–∂–¥–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏. –§–æ—Ä–º—É–ª–∞ –æ–±–Ω–æ–≤–ª–µ–Ω–∏—è –¥–ª—è $\theta$ –Ω–∞ –∫–∞–∂–¥–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏ $t$ –±—É–¥–µ—Ç —Å–ª–µ–¥—É—é—â–µ–π:

$$\theta_{t+1} = \theta_t - \alpha \frac{\partial L(\theta_t)}{\partial \theta}$$

–ú—ã —É–∂–µ –≤—ã—á–∏—Å–ª–∏–ª–∏ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é $L(\theta)$ –≤ –ø—É–Ω–∫—Ç–µ 1. –ü–æ–¥—Å—Ç–∞–≤–∏–º –µ–µ –≤ —Ñ–æ—Ä–º—É–ª—É –æ–±–Ω–æ–≤–ª–µ–Ω–∏—è:

$$\theta_{t+1} = \theta_t - \alpha \sum_{i=1}^{n} (Y_i - \theta_t x_i)(-x_i) = \theta_t + \alpha \sum_{i=1}^{n} (Y_i - \theta_t x_i)x_i$$
$$\theta_{t+1} = \theta_t + \alpha \sum_{i=1}^{n} (Y_i x_i - \theta_t x_i^2)$$

### –°—Ç–æ—Ö–∞—Å—Ç–∏—á–µ—Å–∫–∏–π –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π —Å–ø—É—Å–∫ (SGD):

SGD –∏—Å–ø–æ–ª—å–∑—É–µ—Ç –Ω–µ–±–æ–ª—å—à–æ–π –±–∞—Ç—á –æ–±—ä–µ–∫—Ç–æ–≤ –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –≥—Ä–∞–¥–∏–µ–Ω—Ç–∞ –Ω–∞ –∫–∞–∂–¥–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏. –≠—Ç–æ –ø–æ–∑–≤–æ–ª—è–µ—Ç –±—ã—Å—Ç—Ä–µ–µ —Å—Ö–æ–¥–∏—Ç—å—Å—è –∏ —Å–Ω–∏–∂–∞–µ—Ç –≤–ª–∏—è–Ω–∏–µ –æ—Ç–¥–µ–ª—å–Ω—ã—Ö –≤—ã–±—Ä–æ—Å–æ–≤ –ø–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—é —Å GD.

–û–ø—Ä–µ–¥–µ–ª–∏–º *–±–∞—Ç—á* –∫–∞–∫ —Å–ª—É—á–∞–π–Ω—ã–π –Ω–∞–±–æ—Ä –∏–Ω–¥–µ–∫—Å–æ–≤ $B_t \subset \{1, ..., n\}$. –†–∞–∑–º–µ—Ä –±–∞—Ç—á–∞ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω –∏ —Ä–∞–≤–µ–Ω $m$, –≥–¥–µ $1 \le m \le n$.

–¢–æ–≥–¥–∞ —Ñ–æ—Ä–º—É–ª–∞ –æ–±–Ω–æ–≤–ª–µ–Ω–∏—è –¥–ª—è $\theta$ –Ω–∞ –∫–∞–∂–¥–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏ $t$ –±—É–¥–µ—Ç —Å–ª–µ–¥—É—é—â–µ–π:

$$\theta_{t+1} = \theta_t - \alpha \frac{1}{m} \sum_{i \in B_t} \nabla_{\theta} L_i(\theta_t)$$

–≥–¥–µ:

* $B_t$ - —Å–ª—É—á–∞–π–Ω–æ –≤—ã–±—Ä–∞–Ω–Ω—ã–π –Ω–∞–±–æ—Ä –∏–Ω–¥–µ–∫—Å–æ–≤ —Ä–∞–∑–º–µ—Ä–∞ $m$ –∏–∑ $\{1, ..., n\}$
* $m$ - —Ä–∞–∑–º–µ—Ä –±–∞—Ç—á–∞
* $\nabla_{\theta} L_i(\theta_t)$ - –≥—Ä–∞–¥–∏–µ–Ω—Ç —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ—Ç–µ—Ä—å, –≤—ã—á–∏—Å–ª–µ–Ω–Ω—ã–π —Ç–æ–ª—å–∫–æ –ø–æ *i*-—Ç–æ–º—É –æ–±—ä–µ–∫—Ç—É:
    $$\nabla_{\theta} L_i(\theta_t) = (Y_i - \theta_t x_i)(-x_i) = -(Y_i - \theta_t x_i)x_i$$

–°–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ:

$$\theta_{t+1} = \theta_t + \frac{\alpha}{m} \sum_{i \in B_t} (Y_i - \theta_t x_i)x_i$$

## 3. –í–ª–∏—è–Ω–∏–µ –±–æ–ª—å—à–æ–π –æ—à–∏–±–∫–∏ $\varepsilon_i$

–ï—Å–ª–∏ –¥–ª—è –Ω–µ–∫–æ—Ç–æ—Ä–æ–≥–æ $i$ –æ—à–∏–±–∫–∞ $\varepsilon_i$ –±—É–¥–µ—Ç –æ—á–µ–Ω—å –±–æ–ª—å—à–æ–π, —ç—Ç–æ –ø–æ–≤–ª–∏—è–µ—Ç –Ω–∞ –æ—Ü–µ–Ω–∫—É –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $\theta$, –ø–æ–ª—É—á–µ–Ω–Ω—É—é –∫–∞–∫ –ú–ù–ö, —Ç–∞–∫ –∏ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–º–∏ –º–µ—Ç–æ–¥–∞–º–∏.

* **–ú–ù–ö:** –ü–æ—Å–∫–æ–ª—å–∫—É $\widehat{\theta}$ –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è –Ω–∞ –æ—Å–Ω–æ–≤–µ —Å—É–º–º—ã –ø–æ *–≤—Å–µ–º* –æ–±—ä–µ–∫—Ç–∞–º, –±–æ–ª—å—à–∞—è –æ—à–∏–±–∫–∞ $\varepsilon_i$ –¥–ª—è –æ–¥–Ω–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞ –∏—Å–∫–∞–∑–∏—Ç –æ—Ü–µ–Ω–∫—É $\widehat{\theta}$, "—Å–¥–≤–∏–≥–∞—è" –µ–µ –≤ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–∏ —ç—Ç–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞. –ú–ù–ö —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª–µ–Ω –∫ –≤—ã–±—Ä–æ—Å–∞–º.

* **GD:** –ö–∞–∫ –∏ –≤ —Å–ª—É—á–∞–µ –ú–ù–ö, –±–æ–ª—å—à–∞—è –æ—à–∏–±–∫–∞ –¥–ª—è –Ω–µ–∫–æ—Ç–æ—Ä–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞ –ø–æ–≤–ª–∏—è–µ—Ç –Ω–∞ –≥—Ä–∞–¥–∏–µ–Ω—Ç, –≤—ã—á–∏—Å–ª—è–µ–º—ã–π –Ω–∞ –∫–∞–∂–¥–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏. –û–¥–Ω–∞–∫–æ, —Ç–∞–∫ –∫–∞–∫ GD –∏—Å–ø–æ–ª—å–∑—É–µ—Ç –≤—Å–µ –æ–±—ä–µ–∫—Ç—ã –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –≥—Ä–∞–¥–∏–µ–Ω—Ç–∞, –≤–ª–∏—è–Ω–∏–µ –æ–¥–Ω–æ–π –±–æ–ª—å—à–æ–π –æ—à–∏–±–∫–∏ –±—É–¥–µ—Ç "—Ä–∞–∑–º—ã—Ç–æ" –ø–æ –≤—Å–µ–º –æ–±—ä–µ–∫—Ç–∞–º.

* **SGD:** –ï—Å–ª–∏ –±–æ–ª—å—à–∞—è –æ—à–∏–±–∫–∞ –≤—Å—Ç—Ä–µ—Ç–∏—Ç—Å—è –≤ SGD, —Ç–æ –Ω–∞ *–æ–¥–Ω–æ–π* –∏—Ç–µ—Ä–∞—Ü–∏–∏ –æ–Ω–∞ –æ–∫–∞–∂–µ—Ç –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ–µ –≤–ª–∏—è–Ω–∏–µ –Ω–∞ –æ–±–Ω–æ–≤–ª–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $\theta$. –û–¥–Ω–∞–∫–æ, –ø–æ—Å–∫–æ–ª—å–∫—É SGD –∏—Å–ø–æ–ª—å–∑—É–µ—Ç —Å–ª—É—á–∞–π–Ω—ã–µ –æ–±—ä–µ–∫—Ç—ã –Ω–∞ –∫–∞–∂–¥–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏, —Ç–æ –≤–ª–∏—è–Ω–∏–µ —ç—Ç–æ–π –æ—à–∏–±–∫–∏ –±—É–¥–µ—Ç –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–æ —ç—Ç–æ–π –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–µ–π –∏, –≤–æ–∑–º–æ–∂–Ω–æ, –Ω–µ—Å–∫–æ–ª—å–∫–∏–º–∏ —Å–ª–µ–¥—É—é—â–∏–º–∏, –µ—Å–ª–∏ —ç—Ç–æ—Ç –æ–±—ä–µ–∫—Ç —Å–Ω–æ–≤–∞ –±—É–¥–µ—Ç –≤—ã–±—Ä–∞–Ω. SGD –±–æ–ª–µ–µ —É—Å—Ç–æ–π—á–∏–≤ –∫ –≤—ã–±—Ä–æ—Å–∞–º, —á–µ–º GD –∏ –ú–ù–ö, –Ω–æ –≤—Å–µ —Ä–∞–≤–Ω–æ –ø–æ–¥–≤–µ—Ä–∂–µ–Ω –∏—Ö –≤–ª–∏—è–Ω–∏—é.

–í –æ–±—â–µ–º, –±–æ–ª—å—à–∞—è –æ—à–∏–±–∫–∞ $\varepsilon_i$ –∏—Å–∫–∞–∂–∞–µ—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã, –Ω–æ –≤–ª–∏—è–Ω–∏–µ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç –º–µ—Ç–æ–¥–∞ –æ–±—É—á–µ–Ω–∏—è. –ú–ù–ö –Ω–∞–∏–±–æ–ª–µ–µ —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª–µ–Ω –∫ –≤—ã–±—Ä–æ—Å–∞–º, GD –º–µ–Ω–µ–µ, –∞ SGD –≤ –Ω–µ–∫–æ—Ç–æ—Ä–æ–π —Å—Ç–µ–ø–µ–Ω–∏ —É—Å—Ç–æ–π—á–∏–≤ –∏–∑-–∑–∞ —Å—Ç–æ—Ö–∞—Å—Ç–∏—á–µ—Å–∫–æ–π –ø—Ä–∏—Ä–æ–¥—ã –∞–ª–≥–æ—Ä–∏—Ç–º–∞.

* `season`: –≤—Ä–µ–º—è –≥–æ–¥–∞: 1 &mdash; –∑–∏–º–∞, 2 &mdash; –≤–µ—Å–Ω–∞, 3 &mdash; –ª–µ—Ç–æ, 4 &mdash; –æ—Å–µ–Ω—å;
* `mnth`: –º–µ—Å—è—Ü –æ—Ç 1 –¥–æ 12;
* `holiday`: —è–≤–ª—è–µ—Ç—Å—è –ª–∏ —Ç–µ–∫—É—â–∏–π –¥–µ–Ω—å –ø—Ä–∞–∑–¥–Ω–∏—á–Ω—ã–º;
* `weekday`: –¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏ –æ—Ç 0 –¥–æ 6;
* `workingday`: —è–≤–ª—è–µ—Ç—Å—è –ª–∏ —Ç–µ–∫—É—â–∏–π –¥–µ–Ω—å —Ä–∞–±–æ—á–∏–º –∏–ª–∏ –≤—ã—Ö–æ–¥–Ω—ã–º;
* `weathersit`: –æ—Ü–µ–Ω–∫–∞ –±–ª–∞–≥–æ–ø—Ä–∏—è—Ç–Ω–æ—Å—Ç–∏ –ø–æ–≥–æ–¥—ã –æ—Ç 1 (—á–∏—Å—Ç—ã–π, —è—Å–Ω—ã–π –¥–µ–Ω—å) –¥–æ 4 (–ª–∏–≤–µ–Ω—å, —Ç—É–º–∞–Ω);
* `temp`: —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞ –≤ –¶–µ–ª—å—Å–∏—è—Ö;

* `hum`: –≤–ª–∞–∂–Ω–æ—Å—Ç—å –≤–æ–∑–¥—É—Ö–∞;
* `windspeed`: —Å–∫–æ—Ä–æ—Å—Ç—å –≤–µ—Ç—Ä–∞;
* `cnt`: –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞—Ä–µ–Ω–¥–æ–≤–∞–Ω–Ω—ã—Ö –≤–µ–ª–æ—Å–∏–ø–µ–¥–æ–≤ –≤ —Ç–µ–∫—É—â–∏–π –¥–µ–Ω—å.

–°–∫–∞—á–∞–π—Ç–µ –¥–∞–Ω–Ω—ã–µ, –∑–∞–≥—Ä—É–∑–∏—Ç–µ –∏—Ö —Å –ø–æ–º–æ—â—å—é `pandas` –∏ —Ä–∞–∑–¥–µ–ª–∏—Ç–µ –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é —á–∞—Å—Ç–∏ –≤ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–∏ 4:1. –î–∞–ª–µ–µ –≤—Å—é –∞–Ω–∞–ª–∏—Ç–∏–∫—É –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–æ–≤–æ–¥–∏—Ç—å —Ç–æ–ª—å–∫–æ –Ω–∞ –æ–±—É—á–∞—é—â–µ–π —á–∞—Å—Ç–∏ –¥–∞–Ω–Ω—ã—Ö.

*–£–∫–∞–∑–∞–Ω–∏–µ*. –°—Ä–µ–¥–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –µ—Å—Ç—å —Ç–∞–∫–∏–µ, –∑–Ω–∞—á–µ–Ω–∏—è –∫–æ—Ç–æ—Ä—ã—Ö –∫–∞–∫-—Ç–æ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω—ã, –∏ –∏—Ö –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –Ω–µ–±–æ–ª—å—à–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ. –ß—Ç–æ–±—ã –ø–æ–Ω—è—Ç—å, —è–≤–ª—è—é—Ç—Å—è –ª–∏ –æ–Ω–∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–º–∏, –ø–æ–ø—Ä–æ–±—É–π—Ç–µ –ø–æ–Ω—è—Ç—å, –∫–∞–∫ —Å –∏—Ö –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –±—É–¥—É—Ç —Ä–∞–±–æ—Ç–∞—Ç—å –ª–∏–Ω–µ–π–Ω—ã–µ –º–æ–¥–µ–ª–∏. –ò–Ω–∞—á–µ –≥–æ–≤–æ—Ä—è, –¥–æ–ø—É—Å—Ç–∏–º–æ –ª–∏ —Å —Ç–æ—á–∫–∏ –∑—Ä–µ–Ω–∏—è –ø—Ä–∏—Ä–æ–¥—ã –¥–∞–Ω–Ω—ã—Ö –≤—ã–ø–æ–ª–Ω—è—Ç—å –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∏–µ –æ–ø–µ—Ä–∞—Ü–∏–∏ –º–µ–∂–¥—É *–ª—é–±—ã–º–∏* –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ —ç—Ç–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞, —Å–æ–≥–ª–∞—Å—É–µ—Ç—Å—è –ª–∏ —ç—Ç–æ —Å–æ —Å–º—ã—Å–ª–æ–º, –∫–æ—Ç–æ—Ä—ã–π –Ω–µ—Å–µ—Ç –≤ —Å–µ–±–µ –¥–∞–Ω–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫?

```python
need_features = ['mnth', 'weekday', 'weathersit', 'atemp', 'hum', 'windspeed']

plt.figure(figsize=(14, 8))

for i, column in enumerate(need_features, 1):
    
    plt.subplot (2, 3, i)
    plt.plot(X_train[column], y_train, marker='o', linestyle='', alpha = 0.2, label = column)

plt.show()
```

**a)** —É–±–∏—Ä–∞–µ–º –≤—ã–±—Ä–æ—Å –ø–æ –ø—Ä–∏–∑–Ω–∞–∫—É hum

**b)** –¥–æ–±–∞–≤–ª—è–µ–º –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫ –Ω–∞ –æ—Å–Ω–æ–≤–µ mnth 

**c)** –∫–æ–¥–∏—Ä—É–µ–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –∏ –Ω–æ—Ä–º–∏—Ä—É–µ–º –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏.  

–†–µ–∞–ª–∏–∑—É–π—Ç–µ –¥–∞–Ω–Ω—É—é –ø—Ä–æ—Ü–µ–¥—É—Ä—É.

```python
outlier_index = X_train[X_train['hum'] == 0].index
X_train = X_train.drop(outlier_index)
y_train = y_train.drop(outlier_index)

need_features = ['mnth', 'weekday', 'weathersit', 'atemp', 'hum', 'windspeed']

prepare_X_train = X_train[need_features].copy()
prepare_X_train['mnth^2'] = np.square(prepare_X_train['mnth'])

prepare_X_test = X_test[need_features].copy()
prepare_X_test['mnth^2'] = np.square(prepare_X_test['mnth'])

categorical_cols = ['mnth', 'mnth^2', 'weekday', 'weathersit']
numerical_cols = ['atemp', 'hum', 'windspeed']

encoder = OneHotEncoder(drop="first", sparse_output=False, handle_unknown = 'ignore')

encoder.fit(prepare_X_train[categorical_cols])

X_train_encoded = encoder.transform(prepare_X_train[categorical_cols])
X_test_encoded = encoder.transform(prepare_X_test[categorical_cols])

X_train_encoded_df = pd.DataFrame(
    X_train_encoded, columns=encoder.get_feature_names_out(categorical_cols), index=X_train.index)
X_test_encoded_df = pd.DataFrame(
    X_test_encoded, columns=encoder.get_feature_names_out(categorical_cols), index=X_test.index)

X_train_1 = pd.concat([prepare_X_train.drop(categorical_cols, axis=1), X_train_encoded_df], axis=1)
X_test_1 = pd.concat([prepare_X_test.drop(categorical_cols, axis=1), X_test_encoded_df], axis=1)

X_test_1.head()
```

```python
scaler = StandardScaler()
scaler.fit(X_train_1[numerical_cols])

X_train_2 = X_train_1.copy()
X_train_2[numerical_cols] = scaler.transform(X_train_1[numerical_cols])

X_test_2 = X_test_1.copy()
X_test_2[numerical_cols] = scaler.transform(X_test_1[numerical_cols])
```

<!-- –í–Ω–∏–º–∞–Ω–∏–µ! –ó–¥–µ—Å—å –Ω–∞—Ö–æ–¥–∏—Ç—Å—è —Å–ø–µ—Ü–∏–∞–ª—å–Ω–∞—è –æ—Ç–º–µ—Ç–∫–∞. –ï–µ –Ω–µ–ª—å–∑—è —É–¥–∞–ª—è—Ç—å! –û—Ç–≤–µ—Ç—ã –Ω–∞ –≤–æ–ø—Ä–æ—Å—ã / –≤—ã–≤–æ–¥—ã –¥–ª—è –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –ø—É–Ω–∫—Ç–∞(–æ–≤) –ø–∏—à–∏ –≤ —è—á–µ–π–∫–∞—Ö –≤—ã—à–µ! -->

#### 3. –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ –∏ –µ–µ –∞–Ω–∞–ª–∏–∑

–û–±—É—á–∏—Ç–µ –ª–∏–Ω–µ–π–Ω—É—é —Ä–µ–≥—Ä–µ—Å—Å–∏—é –∏–∑ `sklearn` –Ω–∞ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –∏ –ø–æ—Å–º–æ—Ç—Ä–∏—Ç–µ –Ω–∞ –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –æ—Ü–µ–Ω–∫–∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ (–≤ —Ç.—á. —Å–≤–æ–±–æ–¥–Ω—ã–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç).

```python
model = LinearRegression()
model.fit(X_train_2, y_train)

model.coef_, model.intercept_
```

<!-- –í–Ω–∏–º–∞–Ω–∏–µ! –ó–¥–µ—Å—å –Ω–∞—Ö–æ–¥–∏—Ç—Å—è —Å–ø–µ—Ü–∏–∞–ª—å–Ω–∞—è –æ—Ç–º–µ—Ç–∫–∞. –ï–µ –Ω–µ–ª—å–∑—è —É–¥–∞–ª—è—Ç—å! –û—Ç–≤–µ—Ç—ã –Ω–∞ –≤–æ–ø—Ä–æ—Å—ã / –≤—ã–≤–æ–¥—ã –¥–ª—è –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –ø—É–Ω–∫—Ç–∞(–æ–≤) –ø–∏—à–∏ –≤ —è—á–µ–π–∫–∞—Ö –≤—ã—à–µ! -->

–ü–æ—Å—á–∏—Ç–∞–π—Ç–µ –æ—à–∏–±–∫—É –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ –ø–æ –º–µ—Ç—Ä–∏–∫–∞–º RMSE, MAE, MAPE. –ü–æ–¥—Ä–æ–±–Ω–µ–µ –º–æ–∂–Ω–æ –ø–æ—á–∏—Ç–∞—Ç—å –≤ <a href="https://thetahat.ru/courses/ad/main/3/ad_linreg_sklearn">–Ω–æ—É—Ç–±—É–∫–µ</a> —Å –∑–∞–Ω—è—Ç–∏—è.

*–ó–∞–º–µ—á–∞–Ω–∏–µ.* –ï—Å–ª–∏ –≤ —Ç–µ—Å—Ç–æ–≤–æ–π —á–∞—Å—Ç–∏ —Ç–æ–∂–µ –µ—Å—Ç—å "–±–æ–ª—å—à–∏–µ" –≤—ã–±—Ä–æ—Å—ã, –∏—Ö —Å—Ç–æ–∏—Ç —É–±—Ä–∞—Ç—å. –ù–∞—à–∞ –º–æ–¥–µ–ª—å –Ω–µ –¥–æ–ª–∂–Ω–∞ –ø—ã—Ç–∞—Ç—å—Å—è –∏—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞—Ç—å (–Ω–∞–ø—Ä–∏–º–µ—Ä, —Å–±–æ–∏ –≤ —Ä–∞–±–æ—Ç–µ —Å–µ—Ä–≤–∏—Å–∞) –Ω–æ –≤ —Ç–æ –∂–µ –≤—Ä–µ–º—è –≤—ã–±—Ä–æ—Å—ã –≤–Ω–æ—Å—è—Ç —Å—É—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–π –≤–∫–ª–∞–¥ –≤ –∑–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏. –û—Å–æ–±–µ–Ω–Ω–æ —Å—Ç–æ–∏—Ç –æ–±—Ä–∞—Ç–∏—Ç—å –Ω–∞ —ç—Ç–æ –≤–Ω–∏–º–∞–Ω–∏–µ –≤ —Å–ª—É—á–∞–µ –≤—ã—Å–æ–∫–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –º–µ—Ç—Ä–∏–∫.

*–ü–µ—Ä–µ–¥ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–µ–º –∑–∞–¥–∞—á–∏ –æ–∑–Ω–∞–∫–æ–º—Ç–µ—Å—å —Å <a href="https://thetahat.ru/courses/ad/main/3/ad_logreg_sklearn">–Ω–æ—É—Ç–±—É–∫–æ–º –ø–æ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏</a> —Å –∑–∞–Ω—è—Ç–∏—è.*

> –û–¥–Ω–æ –∏–∑ –∏–Ω—Ç–µ—Ä–µ—Å–Ω—ã—Ö —Å–≤–æ–π—Å—Ç–≤ –º–æ–¥–µ–ª–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ ‚Äî *–ø—Ä–∏ —Å–æ–±–ª—é–¥–µ–Ω–∏–∏ –µ—ë –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏–π* –æ–Ω–∞ –¥–∞–µ—Ç –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å –ø–æ–ª—É—á–∞—Ç—å **–Ω–µ—Å–º–µ—â–µ–Ω–Ω—ã–µ –æ—Ü–µ–Ω–∫–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π** –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç–∏ –æ–±—ä–µ–∫—Ç–∞ –∫ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–º—É –∫–ª–∞—Å—Å—É.  
> –î–ª—è –ø–æ—è—Å–Ω–µ–Ω–∏—è —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º –æ–±—ä–µ–∫—Ç $x$ –∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ $\widehat{p}(x)$ –¥–ª—è –∫–ª–∞—Å—Å–∞ 1. –ï—Å–ª–∏ –≤–∑—è—Ç—å –Ω–µ–±–æ–ª—å—à—É—é –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç—å –æ–±—ä–µ–∫—Ç–∞ $x$, —Ç–æ –¥–æ–ª—è –æ–±—ä–µ–∫—Ç–æ–≤ –∫–ª–∞—Å—Å–∞ 1 –≤ —ç—Ç–æ–π –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏ –±—É–¥–µ—Ç –ø—Ä–∏–±–ª–∏–∑–∏—Ç–µ–ª—å–Ω–æ —Ä–∞–≤–Ω–∞ $\widehat{p}(x)$.  

–î–∞–ª–µ–µ –ø—Ä–æ–≤–µ—Ä–∏–º —ç—Ç–æ —Å–≤–æ–π—Å—Ç–≤–æ –Ω–∞ –∫–æ–Ω–∫—Ä–µ—Ç–Ω—ã—Ö –ø—Ä–∏–º–µ—Ä–∞—Ö.

–° –ø–æ–º–æ—â—å—é –∫–æ–¥–∞ –Ω–∏–∂–µ —Å–≥–µ–Ω–µ—Ä–∏—Ä—É–π—Ç–µ –¥–∞–Ω–Ω—ã–µ, —Å–æ—Å—Ç–æ—è—â–∏–µ –∏–∑ –æ–¥–Ω–æ–≥–æ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –∏ –±–∏–Ω–∞—Ä–Ω–æ–≥–æ —Ç–∞—Ä–≥–µ—Ç–∞.

```python
sample_size = 3000  # –†–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏

# –ü—Ä–∏–∑–Ω–∞–∫–∏
X = np.random.uniform(low=-4, high=4, size=(sample_size, 1))

# –¢–∞—Ä–≥–µ—Ç
y_mean_true = 1 / (1 + np.exp(1 - 2 * X.ravel()))
y = np.random.binomial(n=1, p=y_mean_true)

plt.figure(figsize=(10, 3))
plt.scatter(X, y_mean_true, marker=".", s=1, label="–†–µ–∞–ª—å–Ω—ã–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏\n(–º—ã –∏—Ö –Ω–µ –∑–Ω–∞–µ–º)")
plt.scatter(X, y, marker="|", alpha=0.1, label="–î–∞–Ω–Ω—ã–µ")
plt.xlabel("–ü—Ä–∏–∑–Ω–∞–∫")
plt.ylabel("–ö–ª–∞—Å—Å –æ–±—ä–µ–∫—Ç–∞")
plt.legend();
```

–û–±—É—á–∏—Ç–µ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫—É—é —Ä–µ–≥—Ä–µ—Å—Å–∏—é, –∏—Å–ø–æ–ª—å–∑—É—è —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—é –∏–∑ `sklearn`, –ø—Ä–∏ —ç—Ç–æ–º —Å–≤–æ–±–æ–¥–Ω—ã–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –¥–æ–ª–∂–µ–Ω –ø—Ä–∏—Å—É—Ç—Å—Ç–≤–æ–≤–∞—Ç—å –≤ –º–æ–¥–µ–ª–∏. –£–∫–∞–∂–∏—Ç–µ —Ç–∞–∫–∂–µ `penalty='none'`.

```python
model = LogisticRegression(C=1e9)
model.fit(X, y)
```

–ù–∞–ø–µ—á–∞—Ç–∞–π—Ç–µ –æ—Ü–µ–Ω–∫—É –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤

```python
model.coef_, model.intercept_
```

–ù–∏–∂–µ –æ–±—ä—è–≤–ª–µ–Ω–∞ —Å–µ—Ç–∫–∞ –∑–Ω–∞—á–µ–Ω–∏–π –ø—Ä–∏–∑–Ω–∞–∫–∞. –ü–æ —ç—Ç–æ–π —Å–µ—Ç–∫–µ –ø–æ—Å—Ç—Ä–æ–π—Ç–µ
* –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –∫–ª–∞—Å—Å–æ–≤,
* –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –∫–ª–∞—Å—Å–∞ 1.

–í–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–π—Ç–µ —ç—Ç–∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è. –ù–∞ –≥—Ä–∞—Ñ–∏–∫ —Å—Ç–æ–∏—Ç –Ω–∞–Ω–µ—Å—Ç–∏ —Ç–∞–∫–∂–µ –æ–±—É—á–∞—é—â—É—é –≤—ã–±–æ—Ä–∫—É.

```python
X_grid = np.linspace(-4, 4, 10000).reshape((-1, 1))

y_pred_classes = model.predict(X_grid)
y_pred_proba = model.predict_proba(X_grid)[:, 1]

plt.figure(figsize=(12, 6))
plt.scatter(X, y, color='blue', label='–û–±—É—á–∞—é—â–∞—è –≤—ã–±–æ—Ä–∫–∞', alpha=0.2)
plt.plot(X_grid, y_pred_proba, color='red', label='–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–∞ 1')

plt.plot(X_grid, y_pred_classes, color='green', linestyle='--', label='–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –∫–ª–∞—Å—Å—ã')

plt.xlabel('–ü—Ä–∏–∑–Ω–∞–∫ X')
plt.ylabel('–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –∫–ª–∞—Å—Å–∞ 1 / –ö–ª–∞—Å—Å')
plt.title('–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏')
plt.legend()
plt.grid(True)  # –î–æ–±–∞–≤–ª—è–µ–º —Å–µ—Ç–∫—É –¥–ª—è –ª—É—á—à–µ–π —á–∏—Ç–∞–µ–º–æ—Å—Ç–∏
plt.show()
```

–†–∞–∑–±–µ–π—Ç–µ –æ—Ç—Ä–µ–∑–æ–∫ $[-4, 4]$ –Ω–∞ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ –±–∏–Ω—ã –¥–ª–∏–Ω—ã –¥–ª–∏–Ω—ã 0.2 –∏ –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ –≤ –∫–∞–∂–¥–æ–º –±–∏–Ω–µ –¥–æ–ª—é –æ–±—ä–µ–∫—Ç–æ–≤ –∫–ª–∞—Å—Å–∞ 1. –ü–æ–ª—É—á–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –¥–æ–±–∞–≤—å—Ç–µ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –∏ —Å—Ä–∞–≤–Ω–∏—Ç–µ —ç—Ç–∏ –≥—Ä–∞—Ñ–∏–∫–∏. –ü—Ä–æ–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä—É–π—Ç–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã.

*–ü–æ–¥—Å–∫–∞–∑–∫–∞:* –º–æ–∂–µ—Ç –ø–æ–º–æ—á—å `np.digitize` –∏ –º–µ—Ç–æ–¥ `groupby` –¥–ª—è —Ç–∞–±–ª–∏—Ü `pandas`. –†–µ–∫–æ–º–µ–Ω–¥—É–µ–º –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å <a href="https://thetahat.ru/courses/python">–æ–±—É—á–∞—é—â–∏–µ –Ω–æ—É—Ç–±—É–∫–∏</a> –ø–æ –±–∏–±–ª–∏–æ—Ç–µ–∫–∞–º.

```python
# 1. –†–∞–∑–±–∏–µ–Ω–∏–µ –Ω–∞ –±–∏–Ω—ã —Å –ø–æ–º–æ—â—å—é np.digitize
bins = np.arange(-4, 4 + 0.2, 0.2)  # –°–æ–∑–¥–∞–µ–º –≥—Ä–∞–Ω–∏—Ü—ã –±–∏–Ω–æ–≤
inds = np.digitize(X.ravel(), bins) # –ù–∞—Ö–æ–¥–∏–º –∏–Ω–¥–µ–∫—Å –±–∏–Ω–∞ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è X

# 2. –°–æ–∑–¥–∞–Ω–∏–µ DataFrame –¥–ª—è —É–¥–æ–±—Å—Ç–≤–∞
df = pd.DataFrame({'X': X.ravel(), 'y': y, 'bin': inds})

# 3. –ì—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞ –ø–æ –±–∏–Ω–∞–º –∏ —Ä–∞—Å—á–µ—Ç –¥–æ–ª–∏ –∫–ª–∞—Å—Å–∞ 1
bin_means = df.groupby('bin')['y'].mean()

# 4. –ü–æ–ª—É—á–µ–Ω–∏–µ —Ü–µ–Ω—Ç—Ä–æ–≤ –±–∏–Ω–æ–≤ –¥–ª—è –≥—Ä–∞—Ñ–∏–∫–∞
bin_centers = (bins[:-1] + bins[1:]) / 2 # –í—ã—á–∏—Å–ª—è–µ–º —Å–µ—Ä–µ–¥–∏–Ω—É –∫–∞–∂–¥–æ–≥–æ –±–∏–Ω–∞

# 5. –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è (–¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –Ω–∞ —Å—É—â–µ—Å—Ç–≤—É—é—â–∏–π –≥—Ä–∞—Ñ–∏–∫)
plt.figure(figsize=(12, 6))
plt.scatter(X, y, color='blue', label='–û–±—É—á–∞—é—â–∞—è –≤—ã–±–æ—Ä–∫–∞', alpha=0.2)
plt.plot(X_grid, y_pred_classes, color='green', linestyle='--', label='–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –∫–ª–∞—Å—Å—ã')
plt.plot(bin_centers[bin_means.index-1], bin_means, marker='o', linestyle='-', color='purple', label='–î–æ–ª—è –∫–ª–∞—Å—Å–∞ 1 –≤ –±–∏–Ω–∞—Ö')
plt.plot(X_grid, y_pred_proba, color='red', label='–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–∞ 1')
plt.legend()
plt.show()
```

–ü–æ–≤—Ç–æ—Ä–∏—Ç–µ –ø—Ä–æ–≤–µ–¥–µ–Ω–Ω–æ–µ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ –¥–ª—è —Å–ª–µ–¥—É—é—â–∏—Ö –¥–∞–Ω–Ω—ã—Ö –∏ —Å—Ä–∞–≤–Ω–∏—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã.

```python
# –ü—Ä–∏–∑–Ω–∞–∫–∏
X1 = np.random.uniform(low=-4, high=4, size=(sample_size, 1))

# –¢–∞—Ä–≥–µ—Ç
y1_mean_true = 1 / (1 + np.exp(-100 * X1.ravel()))
y1 = np.random.binomial(n=1, p=y1_mean_true)

plt.figure(figsize=(10, 3))
plt.scatter(X1, y1_mean_true, marker=".", s=1, label="–†–µ–∞–ª—å–Ω—ã–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏")
plt.scatter(X1, y1, marker="|", alpha=0.1, label="–î–∞–Ω–Ω—ã–µ")
plt.xlabel("–ü—Ä–∏–∑–Ω–∞–∫")
plt.ylabel("–ö–ª–∞—Å—Å –æ–±—ä–µ–∫—Ç–∞")
plt.legend();
```

```python
model1 = LogisticRegression(C=1e9)
model1.fit(X1, y1)
y1_pred_classes = model1.predict(X_grid)
y1_pred_proba = model1.predict_proba(X_grid)[:, 1]
```

```python
inds1 = np.digitize(X1.ravel(), bins) # –ù–∞—Ö–æ–¥–∏–º –∏–Ω–¥–µ–∫—Å –±–∏–Ω–∞ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è X

# 2. –°–æ–∑–¥–∞–Ω–∏–µ DataFrame –¥–ª—è —É–¥–æ–±—Å—Ç–≤–∞
df1 = pd.DataFrame({'X': X1.ravel(), 'y': y1, 'bin': inds1})

# 3. –ì—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞ –ø–æ –±–∏–Ω–∞–º –∏ —Ä–∞—Å—á–µ—Ç –¥–æ–ª–∏ –∫–ª–∞—Å—Å–∞ 1
bin1_means = df1.groupby('bin')['y'].mean()

# 4. –ü–æ–ª—É—á–µ–Ω–∏–µ —Ü–µ–Ω—Ç—Ä–æ–≤ –±–∏–Ω–æ–≤ –¥–ª—è –≥—Ä–∞—Ñ–∏–∫–∞
bin_centers = (bins[:-1] + bins[1:]) / 2 # –í—ã—á–∏—Å–ª—è–µ–º —Å–µ—Ä–µ–¥–∏–Ω—É –∫–∞–∂–¥–æ–≥–æ –±–∏–Ω–∞

# 5. –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è (–¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –Ω–∞ —Å—É—â–µ—Å—Ç–≤—É—é—â–∏–π –≥—Ä–∞—Ñ–∏–∫)
plt.figure(figsize=(12, 6))
plt.scatter(X1, y1, color='blue', label='–û–±—É—á–∞—é—â–∞—è –≤—ã–±–æ—Ä–∫–∞', alpha=0.2)
plt.plot(X_grid, y1_pred_classes, color='green', linestyle='--', label='–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –∫–ª–∞—Å—Å—ã')
plt.plot(bin_centers[bin1_means.index-1], bin1_means, marker='o', linestyle='-', color='purple', label='–î–æ–ª—è –∫–ª–∞—Å—Å–∞ 1 –≤ –±–∏–Ω–∞—Ö')
plt.plot(X_grid, y1_pred_proba, color='red', label='–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∫–ª–∞—Å—Å–∞ 1')
plt.legend()
plt.show()
```

**–í—ã–≤–æ–¥—ã:** –ù–∞ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è –ø–æ–∫–∞–∑–∞–ª–∞ —Ö–æ—Ä–æ—à—É—é –ø—Ä–µ–¥—Å–∫–∞–∑–∞—Ç–µ–ª—å–Ω—É—é —Å–ø–æ—Å–æ–±–Ω–æ—Å—Ç—å.

–°–Ω–∞—á–∞–ª–∞ –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ –¥–æ–ª—é –∫–ª–∞—Å—Å–∞ 1 –¥–ª—è –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –æ–±—ä–µ–∫—Ç–æ–≤.

*–ü–æ–¥—Å–∫–∞–∑–∫–∞:* –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ `pd.pivot_table`. –†–µ–∫–æ–º–µ–Ω–¥—É–µ–º –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å <a href="https://thetahat.ru/courses/python">–æ–±—É—á–∞—é—â–∏–µ –Ω–æ—É—Ç–±—É–∫–∏</a> –ø–æ –±–∏–±–ª–∏–æ—Ç–µ–∫–∞–º.

```python
category_proportions = pd.pivot_table(data, values='target', index=["feature_1", "feature_2"] , aggfunc='mean')

print(category_proportions)
```

<!-- –í–Ω–∏–º–∞–Ω–∏–µ! –ó–¥–µ—Å—å –Ω–∞—Ö–æ–¥–∏—Ç—Å—è —Å–ø–µ—Ü–∏–∞–ª—å–Ω–∞—è –æ—Ç–º–µ—Ç–∫–∞. –ï–µ –Ω–µ–ª—å–∑—è —É–¥–∞–ª—è—Ç—å! –û—Ç–≤–µ—Ç—ã –Ω–∞ –≤–æ–ø—Ä–æ—Å—ã / –≤—ã–≤–æ–¥—ã –¥–ª—è –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –ø—É–Ω–∫—Ç–∞(–æ–≤) –ø–∏—à–∏ –≤ —è—á–µ–π–∫–∞—Ö –≤—ã—à–µ! -->
<!-- 14e68d57d58f901b3350e63f76b9f8db -->
–û–±—É—á–∏—Ç–µ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫—É—é —Ä–µ–≥—Ä–µ—Å—Å–∏—é —Å `penalty='none'` –∏ –ø–æ–ª—É—á–∏—Ç–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –¥–ª—è —ç—Ç–∏—Ö —á–µ—Ç—ã—Ä–µ—Ö —Ç–∏–ø–æ–≤ –æ–±—ä–µ–∫—Ç–æ–≤. –ü—Ä–µ–¥—Å—Ç–∞–≤—å—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –≤ —Ç–∞–∫–æ–º –≤–∏–¥–µ, —á—Ç–æ–±—ã –∏—Ö —É–¥–æ–±–Ω–æ –±—ã–ª–æ —Å—Ä–∞–≤–Ω–∏–≤–∞—Ç—å —Å —á–∞—Å—Ç–æ—Ç–∞–º–∏, –ø–æ—Å—á–∏—Ç–∞–Ω–Ω—ã–º–∏ —Ä–∞–Ω–µ–µ.

```python
model2 = LogisticRegression(C=1e9)
model2.fit(data[['feature_1', 'feature_2']], data[['target']])

x_test = pd.DataFrame({
    'feature_1': [0, 0, 1, 1],
    'feature_2': [0, 1, 0, 1]
})

predictions = model2.predict(x_test)
print(predictions)
```
