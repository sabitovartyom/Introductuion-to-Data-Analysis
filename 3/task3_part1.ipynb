Рассмотрим модель одномерной линейной регрессии *без свободного параметра* $y(x) = \theta x$, где $x\in \mathbb{R}$ и $\theta \in \mathbb{R}$. Предполагаем, что данные получены по правилу
$$Y_i = \theta x_i + \varepsilon_i,$$
где $i \in \{1, ..., n\}$, числа $x_i \in \mathbb{R}$ неслучайны, а $\varepsilon_i$ &mdash; случайные ошибки.

**1.** Посчитайте оценку $\widehat{\theta}$ методом наименьших квадратов по явной формуле.

**2.** Выпишите формулы итераций градиентного спуска (GD) и стохастического градиентного спуска (SGD) для поиска $\widehat{\theta}$. Не забудьте указать, как в SGD математически определяется батч (набор объектов на каждой итерации).

**3.** Опишите, что произойдет, если для некоторого $i$ ошибка $\varepsilon_i$ будет очень большой.

**Решение:**

## 1. Оценка $\widehat{\theta}$ методом наименьших квадратов (МНК)

Метод наименьших квадратов стремится минимизировать сумму квадратов ошибок:

$$L(\theta) = \sum_{i=1}^{n} (Y_i - \theta x_i)^2$$

Чтобы найти минимум, возьмем производную по $\theta$ и приравняем ее к нулю:

$$\frac{\partial L(\theta)}{\partial \theta} = \sum_{i=1}^{n} 2(Y_i - \theta x_i)(-x_i) = 0$$


$$\sum_{i=1}^{n} (Y_i x_i - \theta x_i^2) = 0$$


$$\sum_{i=1}^{n} Y_i x_i = \theta \sum_{i=1}^{n} x_i^2$$

Таким образом, оценка $\widehat{\theta}$ методом наименьших квадратов:

$$\widehat{\theta} = \frac{\sum_{i=1}^{n} Y_i x_i}{\sum_{i=1}^{n} x_i^2}$$

## 2. Формулы итераций градиентного спуска (GD) и стохастического градиентного спуска (SGD)

### Градиентный спуск (GD):

GD использует весь набор данных для вычисления градиента на каждой итерации. Формула обновления для $\theta$ на каждой итерации $t$ будет следующей:

$$\theta_{t+1} = \theta_t - \alpha \frac{\partial L(\theta_t)}{\partial \theta}$$

Мы уже вычислили производную $L(\theta)$ в пункте 1. Подставим ее в формулу обновления:

$$\theta_{t+1} = \theta_t - \alpha \sum_{i=1}^{n} (Y_i - \theta_t x_i)(-x_i) = \theta_t + \alpha \sum_{i=1}^{n} (Y_i - \theta_t x_i)x_i$$
$$\theta_{t+1} = \theta_t + \alpha \sum_{i=1}^{n} (Y_i x_i - \theta_t x_i^2)$$

### Стохастический градиентный спуск (SGD):

SGD использует небольшой батч объектов для вычисления градиента на каждой итерации. Это позволяет быстрее сходиться и снижает влияние отдельных выбросов по сравнению с GD.

Определим *батч* как случайный набор индексов $B_t \subset \{1, ..., n\}$. Размер батча фиксирован и равен $m$, где $1 \le m \le n$.

Тогда формула обновления для $\theta$ на каждой итерации $t$ будет следующей:

$$\theta_{t+1} = \theta_t - \alpha \frac{1}{m} \sum_{i \in B_t} \nabla_{\theta} L_i(\theta_t)$$

где:

* $B_t$ - случайно выбранный набор индексов размера $m$ из $\{1, ..., n\}$
* $m$ - размер батча
* $\nabla_{\theta} L_i(\theta_t)$ - градиент функции потерь, вычисленный только по *i*-тому объекту:
    $$\nabla_{\theta} L_i(\theta_t) = (Y_i - \theta_t x_i)(-x_i) = -(Y_i - \theta_t x_i)x_i$$

Следовательно:

$$\theta_{t+1} = \theta_t + \frac{\alpha}{m} \sum_{i \in B_t} (Y_i - \theta_t x_i)x_i$$

## 3. Влияние большой ошибки $\varepsilon_i$

Если для некоторого $i$ ошибка $\varepsilon_i$ будет очень большой, это повлияет на оценку параметра $\theta$, полученную как МНК, так и градиентными методами.

* **МНК:** Поскольку $\widehat{\theta}$ вычисляется на основе суммы по *всем* объектам, большая ошибка $\varepsilon_i$ для одного объекта исказит оценку $\widehat{\theta}$, "сдвигая" ее в направлении этого объекта. МНК чувствителен к выбросам.

* **GD:** Как и в случае МНК, большая ошибка для некоторого объекта повлияет на градиент, вычисляемый на каждой итерации. Однако, так как GD использует все объекты для вычисления градиента, влияние одной большой ошибки будет "размыто" по всем объектам.

* **SGD:** Если большая ошибка встретится в SGD, то на *одной* итерации она окажет значительное влияние на обновление параметра $\theta$. Однако, поскольку SGD использует случайные объекты на каждой итерации, то влияние этой ошибки будет ограничено этой конкретной итерацией и, возможно, несколькими следующими, если этот объект снова будет выбран. SGD более устойчив к выбросам, чем GD и МНК, но все равно подвержен их влиянию.

В общем, большая ошибка $\varepsilon_i$ искажает результаты, но влияние зависит от метода обучения. МНК наиболее чувствителен к выбросам, GD менее, а SGD в некоторой степени устойчив из-за стохастической природы алгоритма.

```python
outlier_index = X_train[X_train['hum'] == 0].index
X_train = X_train.drop(outlier_index)
y_train = y_train.drop(outlier_index)

need_features = ['mnth', 'weekday', 'weathersit', 'atemp', 'hum', 'windspeed']

prepare_X_train = X_train[need_features].copy()
prepare_X_train['mnth^2'] = np.square(prepare_X_train['mnth'])

prepare_X_test = X_test[need_features].copy()
prepare_X_test['mnth^2'] = np.square(prepare_X_test['mnth'])

categorical_cols = ['mnth', 'mnth^2', 'weekday', 'weathersit']
numerical_cols = ['atemp', 'hum', 'windspeed']

encoder = OneHotEncoder(drop="first", sparse_output=False, handle_unknown = 'ignore')

encoder.fit(prepare_X_train[categorical_cols])

X_train_encoded = encoder.transform(prepare_X_train[categorical_cols])
X_test_encoded = encoder.transform(prepare_X_test[categorical_cols])

X_train_encoded_df = pd.DataFrame(
    X_train_encoded, columns=encoder.get_feature_names_out(categorical_cols), index=X_train.index)
X_test_encoded_df = pd.DataFrame(
    X_test_encoded, columns=encoder.get_feature_names_out(categorical_cols), index=X_test.index)

X_train_1 = pd.concat([prepare_X_train.drop(categorical_cols, axis=1), X_train_encoded_df], axis=1)
X_test_1 = pd.concat([prepare_X_test.drop(categorical_cols, axis=1), X_test_encoded_df], axis=1)

X_test_1.head()
```

```python
scaler = StandardScaler()
scaler.fit(X_train_1[numerical_cols])

X_train_2 = X_train_1.copy()
X_train_2[numerical_cols] = scaler.transform(X_train_1[numerical_cols])

X_test_2 = X_test_1.copy()
X_test_2[numerical_cols] = scaler.transform(X_test_1[numerical_cols])
```

### Задача 3.

*Перед выполнением задачи ознакомтесь с <a href="https://thetahat.ru/courses/ad/main/3/ad_logreg_sklearn">ноутбуком по логистической регрессии</a> с занятия.*

> Одно из интересных свойств модели логистической регрессии — *при соблюдении её предположений* она дает возможность получать **несмещенные оценки вероятностей** принадлежности объекта к определенному классу.  
> Для пояснения рассмотрим объект $x$ и соответствующее предсказание вероятности $\widehat{p}(x)$ для класса 1. Если взять небольшую окрестность объекта $x$, то доля объектов класса 1 в этой окрестности будет приблизительно равна $\widehat{p}(x)$.  

Далее проверим это свойство на конкретных примерах.

С помощью кода ниже сгенерируйте данные, состоящие из одного вещественного признака и бинарного таргета.

```python
sample_size = 3000  # Размер выборки

# Признаки
X = np.random.uniform(low=-4, high=4, size=(sample_size, 1))

# Таргет
y_mean_true = 1 / (1 + np.exp(1 - 2 * X.ravel()))
y = np.random.binomial(n=1, p=y_mean_true)

plt.figure(figsize=(10, 3))
plt.scatter(X, y_mean_true, marker=".", s=1, label="Реальные вероятности\n(мы их не знаем)")
plt.scatter(X, y, marker="|", alpha=0.1, label="Данные")
plt.xlabel("Признак")
plt.ylabel("Класс объекта")
plt.legend();
```

Обучите логистическую регрессию, используя реализацию из `sklearn`, при этом свободный коэффициент должен присутствовать в модели. Укажите также `penalty='none'`.

```python
model = LogisticRegression(C=1e9)
model.fit(X, y)
```

Напечатайте оценку коэффициентов

```python
model.coef_, model.intercept_
```

Ниже объявлена сетка значений признака. По этой сетке постройте
* предсказания классов,
* предсказания вероятностей класса 1.

Визуализируйте эти предсказания. На график стоит нанести также обучающую выборку.

```python
X_grid = np.linspace(-4, 4, 10000).reshape((-1, 1))

y_pred_classes = model.predict(X_grid)
y_pred_proba = model.predict_proba(X_grid)[:, 1]

plt.figure(figsize=(12, 6))
plt.scatter(X, y, color='blue', label='Обучающая выборка', alpha=0.2)
plt.plot(X_grid, y_pred_proba, color='red', label='Предсказанные вероятности класса 1')

plt.plot(X_grid, y_pred_classes, color='green', linestyle='--', label='Предсказанные классы')

plt.xlabel('Признак X')
plt.ylabel('Вероятность класса 1 / Класс')
plt.title('Предсказания логистической регрессии')
plt.legend()
plt.grid(True)  # Добавляем сетку для лучшей читаемости
plt.show()
```

Разбейте отрезок $[-4, 4]$ на одинаковые бины длины длины 0.2 и посчитайте в каждом бине долю объектов класса 1. Полученные значения добавьте на график предсказаний вероятностей и сравните эти графики. Проинтерпретируйте полученные результаты.

*Подсказка:* может помочь `np.digitize` и метод `groupby` для таблиц `pandas`. Рекомендуем посмотреть <a href="https://thetahat.ru/courses/python">обучающие ноутбуки</a> по библиотекам.

```python
# 1. Разбиение на бины с помощью np.digitize
bins = np.arange(-4, 4 + 0.2, 0.2)  # Создаем границы бинов
inds = np.digitize(X.ravel(), bins) # Находим индекс бина для каждого значения X

# 2. Создание DataFrame для удобства
df = pd.DataFrame({'X': X.ravel(), 'y': y, 'bin': inds})

# 3. Группировка по бинам и расчет доли класса 1
bin_means = df.groupby('bin')['y'].mean()

# 4. Получение центров бинов для графика
bin_centers = (bins[:-1] + bins[1:]) / 2 # Вычисляем середину каждого бина

# 5. Визуализация (добавление на существующий график)
plt.figure(figsize=(12, 6))
plt.scatter(X, y, color='blue', label='Обучающая выборка', alpha=0.2)
plt.plot(X_grid, y_pred_classes, color='green', linestyle='--', label='Предсказанные классы')
plt.plot(bin_centers[bin_means.index-1], bin_means, marker='o', linestyle='-', color='purple', label='Доля класса 1 в бинах')
plt.plot(X_grid, y_pred_proba, color='red', label='Предсказанные вероятности класса 1')
plt.legend()
plt.show()
```
